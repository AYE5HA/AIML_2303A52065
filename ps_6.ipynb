```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

Generate a normal distribution of data points with a mean of 60 and a standard
deviation of 12. Plot a histogram for the generated data. Analyze the symmetry and
bell-shaped curve of a normal distribution and visualize its Probability Density
Function (PDF).

```python
mean = 60
std_dev = 12
num_samples = 10000
data = np.random.normal(mean, std_dev, num_samples)
```

```python
plt.hist(data, bins=30, density=True, alpha=0.6)
plt.title('Histogram of Normal Distribution')
plt.xlabel('Data Points')
plt.ylabel('Frequency')
plt.show()
```

```python
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
pdf = norm.pdf(x, loc=mean, scale=std_dev)
plt.plot(x, pdf)
plt.title('Probability Density Function (PDF)')
plt.xlabel('Data Points')
plt.ylabel('Probability Density')
plt.show()
```

The normalized distribution for a person’s daily calorie intake has a mean of 30
calories and a standard deviation of 5 calories. Write a python code that Calculates and
visualizes the following probabilities for a randomly selected person having a daily
calorie intake:  
Greater than or equal to 19 calories.  
Less than or equal to 19 calories.  
Between 19 and 35 calories.

```python
mean = 30
std_dev = 5
```

```python
prob_greater_than_19 = 1 - norm.cdf(19, loc=mean, scale=std_dev)
print(f"Probability of calorie intake >= 19: {prob_greater_than_19:.4f}")
```

```python
prob_less_than_19 = norm.cdf(19, loc=mean, scale=std_dev)
print(f"Probability of calorie intake <= 19: {prob_less_than_19:.4f}")
```

```python
prob_between_19_35 = norm.cdf(35, loc=mean, scale=std_dev) - norm.cdf(19, loc=mean, scale=std_dev)
print(f"Probability of calorie intake between 19 and 35: {prob_between_19_35:.4f}")
```

```python
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
pdf = norm.pdf(x, loc=mean, scale=std_dev)

plt.plot(x, pdf, color='blue', label='Normal Distribution')
plt.fill_between(x, pdf, where=(x >= 19), color='red', alpha=0.5, label='>= 19')
plt.fill_between(x, pdf, where=(x <= 19), color='green', alpha=0.5, label='<= 19')
plt.fill_between(x, pdf, where=((x >= 19) & (x <= 35)), color='yellow', alpha=0.5, label='19 to 35')

plt.title('Probability Density Function (PDF) of Daily Calorie Intake')
plt.xlabel('Calorie Intake')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
